```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
df = pd.read_csv("Manual_Log.csv", sep = ";", error_bad_lines=False)
df
```

```python
# Split dataframe into three dataframes based on Server
df = df.drop(['Message_type', 'Transaction_ID','To_Server', 'Message'], axis=1)
df
```

```python
df = df.reset_index()
df = df.drop(['index'], axis=1)
df
```

```python
df["Time"]=df["Time"].div(60)
```

```python
df["Time_floor"] = np.floor(df["Time"]).astype("int")
df.tail()
```

```python
df = df.groupby(['Server','Time_floor'], as_index=False).mean()
df
```

```python
df = df.drop(['Time'], axis=1)
```

```python
df.to_csv("Manual_Log_Filtered.csv")
df.to_excel("Manual_Log_Filtered.xlsx")
```
