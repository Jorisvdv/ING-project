```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Manual_Log.csv", sep = ";", error_bad_lines=False)
```

```python
# Split dataframe into three dataframes based on Server
df_A = df[df['Server'] == "A"]
df_B = df[df['Server'] == "B"]
df_C = df[df['Server'] == "C"]
df_A = df_A.drop(['Message_type', 'Transaction_ID','To_Server', 'Message'], axis=1)
df_B = df_B.drop(['Message_type', 'Transaction_ID','To_Server', 'Message'], axis=1)
df_C = df_C.drop(['Message_type', 'Transaction_ID','To_Server', 'Message'], axis=1)
```

```python
df_A = df_A.reset_index()
df_A = df_A.drop(['index'], axis=1)
df_B = df_B.reset_index()
df_B = df_B.drop(['index'], axis=1)
df_C = df_C.reset_index()
df_C = df_C.drop(['index'], axis=1)
```

```python
df_A["Time"]=df_A["Time"].div(60)
df_B["Time"]=df_B["Time"].div(60)
df_C["Time"]=df_C["Time"].div(60)
```

```python
df_A["Time_floor"] = np.floor(df_A["Time"]).astype("int")
df_B["Time_floor"] = np.floor(df_B["Time"]).astype("int")
df_C["Time_floor"] = np.floor(df_C["Time"]).astype("int")
df_A.tail()
```

```python
df_A = df_A.groupby(['Time_floor'], as_index=False).mean().groupby('Time_floor')['CPU Usage',"Memory Usage"].mean()
df_B = df_B.groupby(['Time_floor'], as_index=False).mean().groupby('Time_floor')['CPU Usage',"Memory Usage"].mean()
df_C = df_C.groupby(['Time_floor'], as_index=False).mean().groupby('Time_floor')['CPU Usage',"Memory Usage"].mean()
```
