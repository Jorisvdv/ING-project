# Test case seasonality

```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 8* np.pi, 2001)
def volume(time, amount, speed_day, offset):
    return amount*(np.sin((x/(speed_day) + offset)) +1.5)
    
plt.plot(x, volume(x,1000,2,3))
plt.xlabel('Time')
plt.ylabel('Number of transactions')
plt.axis('tight')
plt.title("Expected volume per day (Sine wave)")
plt.xlim(0,24)
plt.show()
```

```python
def transaction_time(time, amount, speed_day, offset):
    return 1/np.random.gamma(volume(time,amount,speed_day, offset),1)
plt.plot(x, transaction_time(x,1000,2,4))
plt.xlabel('Time')
plt.axis('tight')
plt.title("Time between incoming transactions")
plt.xlim(0,24)
plt.show()
```

```python
plt.plot(x, 1/transaction_time(x,1000,2,4))
plt.xlabel('Time')
plt.axis('tight')
plt.title("Estimated volume of transactions")
plt.xlim(0,24)
plt.show()
```
