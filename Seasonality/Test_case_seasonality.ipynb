# Test case seasonality

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import os
print(os.getcwd())
from Seasonality_adjustment import Seasonality
csv_filename= "seasonality_values.csv"
seasonality_df = pd.read_csv(csv_filename, sep = ";")
display(seasonality_df.head())
plt.figure(num=1, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(seasonality_df["time"], seasonality_df["scaler_value"])
plt.show()
```

```python
time_scale = np.linspace(0, 600000, 10000)
max_vol = 5000

len(time_scale)
def volume(time, max_volume, csv_file):
    season = Seasonality(csv_file)
    volumes = np.empty(len(time))
    i = 0
    for time_point in time:
        volumes[i] =np.random.gamma(season.scale(time_point) * max_volume,1)
        i += 1
    return(volumes)
    
plt.figure(num=2, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(time_scale/(3600*24), volume(time_scale,max_vol, csv_filename))
plt.xlabel('Time')
plt.ylabel('Number of transactions')
plt.axis('tight')
plt.title("Expected volume per second")
plt.show()

```

```python
plt.figure(num=3, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(time_scale/(3600*24), 1/volume(time_scale,max_vol, csv_filename))
plt.xlabel('Time')
plt.axis('tight')
plt.title("Time between incoming transactions")
plt.show()
```
